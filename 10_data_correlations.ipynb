```python
# ðŸ”— Data Correlations in Pandas

# Correlation is a statistical measure that shows how two numerical variables are related.

# In Pandas, we use `.corr()` to calculate the correlation between columns.
# - Values range from **-1** to **+1**
#   - `+1` = strong positive correlation
#   - `0` = no correlation
#   - `-1` = strong negative correlation
```

```python
import pandas as pd

data = pd.read_csv('data.csv')
df = pd.DataFrame(data)
df
```

```python
## ðŸ§  Calculate Correlation Between All Numeric Columns
df.corr()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

## ðŸ“Š Visualizing Correlation Matrix with Seaborn
# âœ… Use `.corr()` to measure relationships between numeric columns  
# âœ… Use heatmaps to visualize strength and direction of correlations  
# âœ… Useful in feature selection and data analysis
```

```python
# ðŸ“ˆ Pandas - Plotting

# Pandas has built-in support for plotting through the `.plot()` method,  
# powered by **Matplotlib** under the hood.

# You can quickly visualize trends, distributions, and comparisons.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')

df.plot()

plt.show()
```

```python
# Skatter Plot
df = pd.read_csv('data.csv')

df.plot(kind = 'scatter', x = 'Duration', y = 'Calories')

plt.show()
```

```python
df["Duration"].plot(kind = 'hist')

plt.show()
```

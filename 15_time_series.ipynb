```python
# ‚è∞ Working with Time Series in Pandas

# Pandas provides powerful tools for working with **date/time** data,  
# including parsing, indexing, resampling, and time-based filtering.
```

```python
import pandas as pd

data = {
    "Date": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"],
    "Sales": [100, 120, 90, 140, 130]
}

df = pd.DataFrame(data)

# Convert 'Date' column to datetime
df["Date"] = pd.to_datetime(df["Date"])
df
```

```python
## üìÖ Set the datetime column as the DataFrame index
df.set_index("Date", inplace=True)
df
```

```python
## üìà Plot time-based data

import matplotlib.pyplot as plt

df.plot(title="Sales Over Time", marker="o")
plt.ylabel("Sales")
plt.show()
```

```python
## üîç Filter rows using a date range

df["2024-01-02":"2024-01-04"]
```

```python
# Example: Resample daily data into weekly sums

df.resample("W").sum()
```

```python
## üß™ Create a custom date range

pd.date_range(start="2024-01-01", periods=5, freq="D")
```

```python
# ‚úÖ Use `pd.to_datetime()` to convert columns to datetime  
# ‚úÖ Use `set_index()` for time-based indexing  
# ‚úÖ Use slicing, filtering, and `resample()` for powerful time operations
```
